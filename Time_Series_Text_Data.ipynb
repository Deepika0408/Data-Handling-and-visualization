```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data={"date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01"]),
                            "value":[10,15,22,28,35,42]}
df=pd.DataFrame(data)
df.set_index("date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='x')
plt.xlabel("date")
plt.ylabel("value")
plt.title("Time Series With Trend")
plt.grid(True)
plt.show()
```

```python
data={"date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01"]),
                            "value":[10,5,17,15,20,28]}
df=pd.DataFrame(data)
df.set_index("date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["value"],marker='x')
plt.xlabel("date")
plt.ylabel("value")
plt.title("Time Series With Seasonality")
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

text_data="Python is popular programming language and it is very easy to code.Python is a high-level, general-purpose programming language.Its design philosophy emphasizes code readability with the use of significant indentation."
Wordcloud=WordCloud(width=800,height=400,background_color='white').generate(text_data)
plt.figure(figsize=(10,5))
plt.imshow(Wordcloud)
plt.title('Word Cloud')
plt.axis('off')
plt.show()
words=text_data.split()
word_counts=Counter(words)
plt.figure(figsize=(10,5))
plt.bar(word_counts.keys(),word_counts.values())
plt.title('Histogram of words frequency')
plt.xlabel('words')
plt.ylabel('frequency')
plt.xticks(rotation=90)
plt.show
```
